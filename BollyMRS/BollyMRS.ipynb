```python
import pandas as pd
```

```python
df = pd.read_csv("Bollywood-Songs-Dataset(2017-23).csv")
```

```python
df =df.drop('thumbnail',axis=1).reset_index(drop=True)
```

```python
df['lyrics'] =df['lyrics'].str.lower().replace(r'^\w\s',' ').replace(r'^\n','',regex =True)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

def tokenization(txt):
    tokens = nltk.word_tokenize(txt)
    stemming = [stemmer.stem(w) for w in tokens]
    return " ".join(stemming)
```

```python
df['lyrics'] = df['lyrics'].apply(lambda x: tokenization(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidvector = TfidfVectorizer(analyzer='word',stop_words='english')
matrix = tfidvector.fit_transform(df['lyrics'])
similarity = cosine_similarity(matrix)
```

```python
similarity[0]
```

```python
df[df['music_name']=='Hue Bechain']
```

```python
def recommendation(music_name_df):
    idx = df[df['music_name'] == music_name_df].index[0]
    distance = sorted(list(enumerate(similarity[idx])),reverse=True,key=lambda x:x[1])

    songs = []
    for m_id in distance[1:21]:
       songs.append(df.iloc[m_id[0]].music_name)
    return songs
```

```python
recommendation('Hue Bechain')
```

```python
import pickle
pickle.dump(similarity,open('similarity.pkl','wb'))
pickle.dump(df,open('df.pkl','wb'))
```
